```python
def fib (n):
  a=0
  b=1
  for i in range(n):
    print(a)
    f=a+b
    a=b
    b=f
n= int(input("Enter the number of terms: "))
fib(n)
```

```python
def myPow(x, n):
  i=1
  result= x
  while i<n:
    result *= x
    i+=1
  return result
myPow(5,3)
```

```python
def sumDigits(n):
    r = n
    sum = 0
    while r != 0:
        q = r // 10
        digit = r % 10
        sum += digit
        r = q
    return sum

print(sumDigits(234))
```

```python
def squaredSum(n):
  sum = 0
  for i in n:
    sum += i * i
  print(sum)
squaredSum([2,3,4])
```
